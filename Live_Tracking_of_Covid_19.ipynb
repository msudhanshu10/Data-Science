```
import pandas as pd
df= pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```
df.head()
```

```
df=df.iloc[:-1, 1:]
```

```
import matplotlib.pyplot as plt
```

```
import numpy as np
```

```
total_states= np.arange(len(df['state_name']))
total_states
## will be used in plotting
```

```
plt.figure(num=None, figsize=(9,7))
plt.barh(total_states,df['positive'])
plt.yticks(total_states, df['state_name'])
plt.ylabel('State Name')
plt.xlabel('Total Positive Cases Reported')
```

```
plt.figure(num=None, figsize=(9,7))
plt.barh(total_states,df['new_active'])
plt.yticks(total_states, df['state_name'])
plt.ylabel('State Name')
plt.xlabel('Total Active Cases Reported')
plt.show()
```

```
plt.figure(num=None, figsize=(9,7))
plt.barh(total_states,df['death'])
plt.yticks(total_states, df['state_name'])
plt.ylabel('State Name')
plt.xlabel('Total Deaths Reported')
plt.show()
```

```
plt.figure(num=None, figsize=(9,7))
plt.barh(total_states,df['cured'])
plt.yticks(total_states, df['state_name'])
plt.ylabel('State Name')
plt.xlabel('Total Cured Cases Reported')
plt.show()
```
